```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("/content/data.csv", encoding="latin1");
df = df.dropna(subset=["no2"]);
x = df["no2"].values;
```

```python
r = 102483078;
a_r = 0.05 * (r % 7);
b_r = 0.3 * (r % 5 + 1);
```

```python
z = x + a_r * np.sin(b_r * x);
```

```python
mu = np.mean(z);
variance = np.var(z);
lam = 1 / (2 * variance);
c = np.sqrt(lam / np.pi);
```

```python
results = pd.DataFrame({
    "Parameter": ["mu", "lambda", "c"],
    "Value": [mu, lam, c]
});
results;
```

```python
z_range = np.linspace(min(z), max(z), 500);
pdf = c * np.exp(-lam * (z_range - mu)**2);
```

```python
plt.figure(figsize=(8,5));
plt.hist(z, bins=40, density=True);
plt.plot(z_range, pdf);
plt.xlabel("z");
plt.ylabel("Probability Density");
plt.title("Learned Probability Density Function");
plt.show();
```
